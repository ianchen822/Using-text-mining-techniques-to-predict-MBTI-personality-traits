```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mbti_1.csv", encoding='utf-8')
```

### 清理URL和|||分隔線

```python
import re
```

```python
def clean_split(text):
    text = text.split('|||')
    return text
```

```python
def clean_link(lst):
    for i in range(len(lst)):
        lst[i] = re.sub(r'http\S+', r'<URL>', lst[i])
    return " ".join(lst)[1:-1]
```

```python
df['clean'] = df['posts'].apply(clean_split)
df['clean'] = df['clean'].apply(clean_link)
```

```python
df
```

### tokenization + stemming + TF-IDF

```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def token_stem(text):
    tokens = word_tokenize(text)
    stems = [stemmer.stem(item) for item in tokens]
    return " ".join(stems)
    
```

```python
df['clean'] = df['clean'].apply(token_stem)
```

```python
df['clean']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=10000, lowercase=True, stop_words='english')
tfidf_vectors = tfidf_vectorizer.fit_transform(df['clean'])
```

```python
corpus = pd.DataFrame(tfidf_vectors.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
corpus
```

```python
corpus.insert(0, column="label_type", value=df['type'])
```

```python
corpus.to_csv('mbti_tfidf.csv',index=False)
```
